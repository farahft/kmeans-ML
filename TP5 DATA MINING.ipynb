```python
import pandas as pd

iris = pd.read_csv("iris.csv")
iris
```

```python
X=iris[["sepal_length_in_cm","sepal_width_in_cm"]]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
print(kmeans.labels_) # label of clusters got assigned in
print(kmeans.cluster_centers_)
print(kmeans.score(X))
```

```python
import matplotlib.pyplot as plt

plt.scatter(X.sepal_length_in_cm,X.sepal_width_in_cm,c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c="r")
```

```python
kmeans.predict([[0,0],[55,30]])
```

```python
import numpy as np
def distance(o1,o2):
    return np.sum((o1-o2)**2)
```

```python
def inertia(X,cent,lab):
    k=cent.shape[0]
    n=X.shape[0]
    a=0
    for i in range(k):
        for j in range(n):
            if lab[j]==i:
                a+=distance(X.values[j],cent[i])
    return(a)
```
